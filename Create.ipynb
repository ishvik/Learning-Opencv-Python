# HISTOGRAM

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import cv2
```

```python
img = cv2.imread("cvimg.jpg")
```

```python
#make histogram of quantity of color in img
histogram = cv2.calcHist([img],[0],None,[256],[0,256]) ##parameters = input_image, color, full show, size of histogram, range
```

```python
##plot histogram
plt.hist(img.ravel(),256,[0,256]); plt.show() ##it will show only blue color 
```

```python
##show all color
color = ('b','g','r')
for i,col in enumerate(color):
    hist1 = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist1, color = col)
    plt.xlim([0,256])
plt.show()
```


# creating image and draw on it


```python
import numpy as np
import cv2
```

```python
##image with white background
new = np.zeros((512,512),np.uint8)
new.fill(255)
cv2.imshow("White image ",new)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
##image with black background
new1 = np.zeros((512,512,3),np.uint8)
cv2.imshow("Black image ",new1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
##draw line 
cv2.line(new1,(0,0),(156,200),(255,127,0),8) ##parameters :- starting pt, ending pt, color, thickness of line
cv2.imshow("line",new1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
