```python
##converting image to HSV color space 
hsvimg = cv2.cvtColor(img1,cv2.COLOR_BGR2HSV)
cv2.imshow("HSV",hsvimg)
cv2.imshow("Hue",hsvimg[:,:,0])
cv2.imshow("Saturation",hsvimg[:,:,1])
cv2.imshow("value",hsvimg[:,:,2])
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
##seperating BGR color and re-make again with some modifications
img2 = cv2.imread("cvimg2.jpeg")
B, G, R = cv2.split(img2)
# cv2.imshow("Red",R)
# cv2.imshow("Green",G)
# cv2.imshow("Blue",B)

mergedimg = cv2.merge([B,G,R])
# cv2.imshow("Merge",mergedimg)

addblue = cv2.merge([B+120,G,R])
cv2.imshow("Add blue color",addblue)
cv2.imwrite("blue.jpg",addblue) ##saving image of added more blue color
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
##doing the same as above with numpy
zeros = np.zeros(img2.shape[:2], dtype="uint8")

cv2.imshow("Red",cv2.merge([zeros,zeros,R]))
cv2.imshow("green",cv2.merge([zeros,G,zeros]))
cv2.imshow("blue",cv2.merge([B,zeros,zeros]))

cv2.waitKey(0)
cv2.destroyAllWindows()
```
